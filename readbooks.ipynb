```python
import urllib.parse
import requests
from bs4 import BeautifulSoup

def extract_names_and_books_from_wikipedia(url, starting_letter):
    try:
        # Fetch the Wikipedia page
        page = requests.get(url)
        soup = BeautifulSoup(page.content, 'html.parser', from_encoding=page.encoding)

        # Find all links on the Wikipedia page
        links = soup.find_all("a")

        # Display only links with titles starting with the specified letter
        for link in links:
            title = link.get("title")
            if title and title.startswith(starting_letter):
                print(f"Author: {title}")
                show_one_book(title)

    except requests.exceptions.RequestException as e:
        print(f"An error occurred during the request: {e}")

def show_one_book(name):
    try:
        name = urllib.parse.quote_plus(name)
        open_library_url = f'http://openlibrary.org/search.json?author={name}&limit=1'
        response = requests.get(open_library_url)
        data = response.json()

        # Check if there are works for the author
        if 'docs' in data and data['docs']:
            first_book_title = data['docs'][0]['title']
            print(f"First Book Title: {first_book_title}")
        else:
            print(f"No book title found for {name}")

    except Exception as e:
        print(f"An error occurred for {name}: {e}")

# Example usage for authors starting with "Sh"
wikipedia_url = "https://en.m.wikipedia.org/wiki/List_of_authors_by_name:_M"
starting_letter = "Sh"
extract_names_and_books_from_wikipedia(wikipedia_url, starting_letter)

```
